```python
import os
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
import shutil
import time
import random
```

```python
location = '/home/grzegorz/Google-data'
```

```python
files = os.listdir(os.path.join(location, 'images_res'))
labels = pd.read_feather(os.path.join(location, 'labels.feather'))
```

```python
percentage = 0.1
size = int(len(files) * percentage)
```

```python
test_files = random.sample(files, size)
```

```python
landmark = labels.landmark_id.unique()
```

```python
def MakeDirTest(landmark_id):
    if(not os.path.exists(os.path.join(location, 'test', str(landmark_id)))):
        os.makedirs(os.path.join(location, 'test', str(landmark_id)))
```

```python
Parallel(n_jobs = os.cpu_count(), verbose = 10)(delayed(MakeDirTest)(ld_id) for ld_id in landmark)
```

```python
def CategorizeImage(image):
    f1, f2 = str.split(image, '-')
    cat = str((labels[labels.id == f1].landmark_id.tolist())[0])
    shutil.move(os.path.join(location, 'images_res', image), 
                os.path.join(location, 'train', cat, image))
```
